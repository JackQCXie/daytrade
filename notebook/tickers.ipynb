```python
import pandas
import os
import sys
```

```python
url = 'https://www.ssga.com/us/en/intermediary/library-content/products/fund-data/etfs/us/holdings-daily-us-en-spy.xlsx'
df = pandas.read_excel(url, header=4)

df.columns = df.columns.str.lower().str.replace(' ', '')

# remove extra rows after data
cond = df.isna().all(axis=1)
idxmax = cond.idxmax()

cond = df.index < idxmax
df = df[cond]

df = df.sort_values('weight', ascending=False)
df = df.drop_duplicates('ticker')
df = df.reset_index()

print(df)

# save data in chunks
nrows = len(df)
chunk_size = 50

# last index is included
indices = list(range(0, nrows, chunk_size))
indices[-1] = nrows

# save chunks
n = 0
for start, end in zip(indices[:-1], indices[1:]):
    
    chunk = df.iloc[start:end]

    n+=1
    dst = os.path.abspath(f'../data/spy/spy-{n}.csv')
    chunk.to_csv(dst, index=False)
    print('Saved:', dst)
```

```python
url = 'https://www.ishares.com/us/products/239707/ishares-russell-1000-etf/1467271812596.ajax?fileType=csv&fileName=IWB_holdings&dataType=fund'
df = pandas.read_csv(url, skiprows=9)

df.columns = df.columns.str.lower().str.replace(' ', '')
df.columns = map(lambda x: ''.join(x for x in x if x.isalpha()), df.columns)

# remove extra rows after data
cond = ~df['name'].isna()
df = df[cond]

df = df.sort_values('weight', ascending=False)
df = df.drop_duplicates('ticker')
df = df.reset_index()

print(df)

# save data in chunks
nrows = len(df)
chunk_size = 50

# last index is included
indices = list(range(0, nrows, chunk_size))
indices[-1] = nrows

# save chunks
n = 0
for start, end in zip(indices[:-1], indices[1:]):
    
    chunk = df.iloc[start:end]

    n+=1
    dst = os.path.abspath(f'../data/iwb/iwb-{n}.csv')
    chunk.to_csv(dst, index=False)
    print('Saved:', dst)
```
