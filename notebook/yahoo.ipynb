```python
# if __name__ == '__main__':

            
#     # parse results

symbol = 'aapl'

def clean_price(symbol):

    summary = get_summary(symbol)

    chart = get_chart(symbol)

    result = chart['chart']['result'][0]

    timestamp = result['timestamp']
    quote = result['indicators']['quote'][0]
    adjclose = result['indicators']['adjclose'][0]
    data = quote | adjclose

    df = pandas.DataFrame(data, index=timestamp)
    df.index = df.index.astype('datetime64[s]').date
    df = df.reset_index(names='date')

    # adjusted price adjusts for dividends also
    df['adj'] = df['adjclose'] / df['close']


    cols = ['open', 'close', 'low', 'high', 'volume']
    df[cols] = df[cols].mul(df['adj'], axis=0)

    # mean price
    cols = ['open', 'close', 'low', 'high']
    df['mean'] = df[cols].mean(axis=1)

    # volume by outstanding shares
    shares_outstanding = summary['quoteSummary']['result'][0]['defaultKeyStatistics']['sharesOutstanding']['raw']
    df['volshare'] = df['volume'] / shares_outstanding

    # intraday mean price market cap
    df['mkcap'] = df['mean'] * shares_outstanding
    df = df.dropna(axis=0, how='all')

    return df

    cols = ['open', 'close', 'low', 'high', 'mean', 'volume']
    dfr = df[cols].pct_change()
    dfr = dfr.dropna(axis=0, how='all')
```

```python
def clean_earnings(symbol):

    data = get_earnings(symbol)
    earn = pandas.DataFrame(data)
    ds = earn['Earnings Date'].str.split(' at ').str[0]
    earn['date'] = pandas.to_datetime(ds).dt.date

    # clean columns
    cols = earn.columns
    cols = cols.str.lower()
    remove = ['(%)', 'eps']
    for x in remove:
        cols = cols.str.replace(x, '')
    cols = cols.str.strip()
    cols = cols.str.replace(' ', '_')
    earn.columns = cols

    # clean data
    cols = ['estimate', 'reported', 'surprise']
    earn[cols] = earn[cols].replace('-', 'NaN').astype(float)
    earn['surprise'] = earn['surprise'] * 0.01
    
    return earn
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot') # 
dcolors = plt.rcParams['axes.prop_cycle'].by_key()['color']

%matplotlib inline
```

```python
import os
import json

tickers = []
data_dir = os.path.abspath('../data')
spy_dir = os.path.join(data_dir, 'spy')


files = sorted(os.listdir(spy_dir))
for file in files:
    src = os.path.join(spy_dir, file)
    spy = pandas.read_csv(src, index_col=0)
    tickers.extend(spy['ticker'].tolist())
```

```python
DAYS = 30
N = 20

N = min(len(tickers), N)
tickers = tickers[:N]
dfs = []
t0 = datetime.now()

# plot relative mean price around earnings calls
for n, ticker in enumerate(tickers):
    
    print(f'[{n+1} of {N}] {ticker}')


    symbol = ticker.replace('.', '-')
    dfp = clean_price(symbol)

    cols = ['open', 'close', 'low', 'high', 'mean', 'volume']
    dfr = dfp[cols].pct_change().dropna(how='all')
    dfe = clean_earnings(symbol)

    cols = ['date', 'symbol', 'estimate', 'reported', 'surprise']
    df = pandas.merge(dfp, dfe[cols], on='date', how='left')

    # forward fill earnings date
    cond = ~df['surprise'].isna()
    df.loc[cond, 'earn_date'] = df.loc[cond, 'date']
    df.loc[cond, 'rel'] = df.loc[cond, 'mean']

    fill_cols = ['earn_date', 'rel', 'symbol', 'estimate', 'reported', 'surprise']
    df[fill_cols] = df[fill_cols].ffill()

    # calculate days from earning
    df['days'] = df['date']-df['earn_date']
    df['days'] = df['days'].map(lambda x: x.days if not pandas.isna(x) else x)

    # back fill earnings date
    cond = (abs(df['days']) > DAYS)
    df.loc[cond, 'earn_date'] = None
    df.loc[cond, 'rel'] = None

    df[fill_cols] = df[fill_cols].bfill()
    df['days'] = df['date']-df['earn_date']
    df['days'] = df['days'].map(lambda x: x.days if not pandas.isna(x) else x)
    df['rel'] = (df['mean'] / df['rel']) - 1

    # return only dates within
    cond = (abs(df['days']) <= DAYS)
    df = df[cond]
    dfs.append(df)

    now = datetime.now()
    lap = now - t0
    avg = lap / (n+1)
    eta = avg * (N-n-1)
    print(f'now: {now} | eta: {eta} | lap: {lap} | avg: {avg}')
```

```python

sample = pandas.concat(dfs)
pvt = sample.pivot_table(index='days', values='rel', aggfunc=('mean', 'std'))

# confidence intervals
pvt['upper'] = pvt['mean'] + pvt['std'] * 1.96
pvt['lower'] = pvt['mean'] - pvt['std'] * 1.96

pvt = pvt.reset_index()

fig, ax = plt.subplots()

x, y = pvt['days'], pvt['mean']
u, l = pvt['upper'], pvt['lower']
ax.plot(x, y)
ax.fill_between(x, l, u, color=dcolors[0], alpha=.1)


ax.set_title(f'Relative price earnings calls')
ax.axhline(0, lw=0.5)
c = 'gray'
ax.axvline(0, c=c, ls='--')
ax.set_xlabel('Days after earnings call')
ax.set_xlim(-30, 30)

plt.show()
```
